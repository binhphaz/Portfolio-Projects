```python
#Import Pandas Numpy
import pandas as pd
import numpy as np
```

```python
#Read DataFrame Sales vs 
sales=pd.read_csv("sales.csv")
prices=pd.read_csv("prices.csv")
```

```python
#convert to datetime object
sales["ordered_at"] = pd.to_datetime(sales["ordered_at"])
prices["updated_at"] = pd.to_datetime(prices["updated_at"])
```

```python
#closet_date_function
def closet_update_date (id, order_date):
    list_date=prices["updated_at"][prices["product_id"]==id]
    if order_date<list_date.min():
        closet_date="old"
    else:
        closet_date=list_date[list_date<=order_date].max()
    return closet_date
```

```python
#get_closet_date
for i in np.arange(len(sales)):
    updated_dated=closet_update_date(sales.loc[i]["product_id"],sales.loc[i]["ordered_at"])
    print(updated_dated)
```

```python
#creat_DataFrame_closet_date
row=[]
for i in np.arange(len(sales)):
    updated_dated=closet_update_date(sales.loc[i]["product_id"],sales.loc[i]["ordered_at"])
    row.append([updated_dated])
date_update = pd.DataFrame(row, columns=["A"])
print(date_update)
```

```python
#add date_update to sales
sales["date_update"]=date_update
```

```python
#Covert date to the same data type
sales["date_update"]=sales["date_update"].astype(str)
prices["updated_at"]=prices["updated_at"].astype(str)
```

```python
#Join sales vs prices to get the Final_Price_Update
Final_Price_Update=pd.merge(sales, prices, how="left", left_on=["product_id","date_update"], right_on=["product_id","updated_at"])
```

```python
Final_Price_Update=Final_Price_Update.fillna(60000)
```

```python
Final_Price_Update["revenue"]=Final_Price_Update["new_price"]*Final_Price_Update["quantity_ordered"]
```
